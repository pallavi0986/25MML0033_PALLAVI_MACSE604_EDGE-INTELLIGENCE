```python
import kagglehub
import os
from PIL import Image
import matplotlib.pyplot as plt

path = kagglehub.dataset_download("dataclusterlabs/indian-number-plates-dataset")
print("Path to dataset files:", path)


image_extensions = ('.jpg', '.jpeg', '.png')
all_images = []
for root, dirs, files in os.walk(path):
    for file in files:
        if file.lower().endswith(image_extensions):
            all_images.append(os.path.join(root, file))

if all_images:
    plt.figure(figsize=(12, 8))

    for i in range(min(3, len(all_images))):
        img = Image.open(all_images[i])
        plt.subplot(1, 3, i + 1)
        plt.imshow(img)
        plt.title(f"Sample {i+1}")
        plt.axis('off')
    plt.show()
else:
    print("No images found in the path. Please verify the download.")
```

```python
from PIL import Image
import matplotlib.pyplot as plt

if all_images:

    img_path = all_images[0]
    original_img = Image.open(img_path)

    new_size = (200, 200)
    resized_img = original_img.resize(new_size)

    plt.figure(figsize=(10, 5))

    plt.subplot(1, 2, 1)
    plt.imshow(original_img)
    plt.title(f"Original Image ({original_img.width}x{original_img.height})")
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(resized_img)
    plt.title(f"Resized Image ({resized_img.width}x{resized_img.height})")
    plt.axis('off')

    plt.show()
else:
    print("No images available to resize.")
```

```python
import cv2

print("cv2 library imported successfully.")
```

```python
import matplotlib.pyplot as plt

if all_images:
    # Load the first image using cv2.imread
    img_path = all_images[0]
    original_img_cv2 = cv2.imread(img_path)

    # Check if image was loaded successfully
    if original_img_cv2 is None:
        print(f"Error: Could not load image at {img_path}")
    else:

        gray_img = cv2.cvtColor(original_img_cv2, cv2.COLOR_BGR2GRAY)


        thresh_img = cv2.adaptiveThreshold(gray_img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2)

        plt.figure(figsize=(15, 5))

        plt.subplot(1, 3, 1)

        plt.imshow(cv2.cvtColor(original_img_cv2, cv2.COLOR_BGR2RGB))
        plt.title('Original Image')
        plt.axis('off')

        plt.subplot(1, 3, 2)
        plt.imshow(gray_img, cmap='gray')
        plt.title('Grayscale Image')
        plt.axis('off')

        plt.subplot(1, 3, 3)
        plt.imshow(thresh_img, cmap='gray')
        plt.title('Adaptive Thresholded Image')
        plt.axis('off')

        plt.show()
else:
    print("No images available to process.")
```

```python
from difflib import SequenceMatcher


is_match = (cleaned_extracted_text == cleaned_ground_truth_text)


similarity_ratio = SequenceMatcher(None, cleaned_ground_truth_text, cleaned_extracted_text).ratio()


print(f"\nComparison Result:")
print(f"Ground Truth: {cleaned_ground_truth_text}")
print(f"Extracted:    {cleaned_extracted_text}")
print(f"Exact Match:  {is_match}")
print(f"Similarity Ratio: {similarity_ratio:.2f}")
```
