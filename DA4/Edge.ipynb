```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
import pickle
import os

# 1st data jar
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()

# Normalize the data
x_train = x_train.astype("float32") / 255.0
x_test = x_test.astype("float32") / 255.0

# Reshape the data
x_train = x_train.reshape((x_train.shape[0], 28 * 28))
x_test = x_test.reshape((x_test.shape[0], 28 * 28))

# Model
model = keras.Sequential([
    layers.Dense(128, activation='relu', input_shape=(784,)),
    layers.Dense(10, activation='softmax')
])

# Compile the model
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

print("Training the model...")
model.fit(
    x_train,
    y_train,
    batch_size=32,
    epochs=5, 
    validation_split=0.2
)

# --- Save the model using pickle ---
model_filename = 'keras_mnist_model.pkl'

try:
    with open(model_filename, 'wb') as f:
        pickle.dump(model, f)
    print(f"\nModel successfully saved to '{model_filename}' using pickle.")
    print(f"File Size: {os.path.getsize(model_filename) / 1024:.2f} KB")

except Exception as e:
    print(f"\nError saving model with pickle: {e}")
    print("Keras models often have issues with pickle due to complex internal structures.")
    print("Consider using model.save() instead.")
```
